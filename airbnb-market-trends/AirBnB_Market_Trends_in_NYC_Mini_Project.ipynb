## Loading & Examining Data

```python
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from io import StringIO
from io import BytesIO
```

```python
#Load CSV file

#read in the CSV data from a Google Drive shareable link
csv_orig_url="https://drive.google.com/file/d/1J2isp3aAIBSyGkEgAif3MuomfxXwlzdm/view?usp=sharing"
csv_file_id = csv_orig_url.split('/')[-2] #extract file ID from url
csv_dwn_url = 'https://drive.google.com/uc?export=download&id=' + csv_file_id

#download and convert CSV content into a readable format
csv_url = requests.get(csv_dwn_url).text
csv_raw = StringIO(csv_url)

#read the data into a pandas DataFrame
df_price = pd.read_csv(csv_raw)

#examine first 5 rows of CSV price data
df_price.head()
```

```python
#check price columns and data types
df_price.info()
```

```python
#summary statistics for price numeric columns
df_price.describe()
```

```python
#Load Excel file

#read in the Excel data from a Google Drive shareable link
excel_orig_url="https://docs.google.com/spreadsheets/d/1sqlyXgd1J6CjHlIsd4oYHeLNjnBcJreE/edit?usp=sharing&ouid=113904096852535515268&rtpof=true&sd=true"
excel_file_id = excel_orig_url.split('/')[-2] #extract file ID from url
excel_dwn_url = 'https://drive.google.com/uc?export=download&id=' + excel_file_id

#download the file content as bytes
excel_url = requests.get(excel_dwn_url)
excel_raw = BytesIO(excel_url.content)

#read the data into a pandas DataFrame
df_room_type = pd.read_excel(excel_raw)

#examine first 5 rows of Excel room_type data
df_room_type.head()
```

```python
#check room_type columns and data types
df_room_type.info()
```

```python
#examine number of unique room_type values
print(df_room_type['room_type'].unique())
```

```python
#Load TSV file

#read in the TSV data from a Google Drive shareable link
tsv_orig_url="https://drive.google.com/file/d/1jrjGrzVeqeFFlgZ7HVoWKgRlmGtm3Um7/view?usp=sharing"
tsv_file_id = tsv_orig_url.split('/')[-2] #extract file ID from url
tsv_dwn_url = 'https://drive.google.com/uc?export=download&id=' + tsv_file_id

#download and convert TSV content into a readable format
tsv_url = requests.get(tsv_dwn_url).text
tsv_raw = StringIO(tsv_url)

#read the data into a pandas DataFrame
df_last_review = pd.read_csv(tsv_raw, sep='\t')

#examine first 5 rows of TSV last_review data
df_last_review.head()
```

```python
#check last_review column and data types
df_last_review.info()
```

## Merging the 3 Different DataFrames

```python
#since each DataFrame shares the column "listing id," I will use this to merge them

#first merge df_price and df_room_type on "listing_id"
merged_df = pd.merge(df_price, df_room_type, on='listing_id')

#then merge the result with df_last_review on "listing_id"
merged_df = pd.merge(merged_df, df_last_review, on='listing_id')

#check the merged DataFrame
print(merged_df.head())
```

## Determining the Earliest and Most Recent Review Dates

```python
#convert relevant reviews column to date format
merged_df['last_review'] = pd.to_datetime(merged_df['last_review'], errors='coerce')

#find the earliest review date
first_reviewed = merged_df['last_review'].min()

#find the latest review date
last_reviewed = merged_df['last_review'].max()

#print the results
print(f"Earliest review date: {first_reviewed}")
print(f"Most recent review date: {last_reviewed}")
```

## Finding How Many Listings are Private Rooms

```python
#convert all room_type values to lowercase to standardize them
merged_df['room_type'] = merged_df['room_type'].str.lower()

#filter the DataFrame for rows where room_type is "private room"
private_rooms_df = merged_df[merged_df['room_type'] == 'private room']

#count how many listings are private rooms
nb_private_rooms = private_rooms_df.shape[0]

#print the result
print(f"Number of private room listings: {nb_private_rooms}")
```

## Finding the Average Price of Listings

```python
#remove text/symbols like "dollars" or "$" using str.replace to standardize prices
merged_df['price_cleaned'] = merged_df['price'].str.replace('[^\d.]', '', regex=True) #The regex pattern '[^\d.]' removes any character that's not a digit or dot

#convert the cleaned string values to float
merged_df['price_cleaned'] = merged_df['price_cleaned'].astype(float)

#calculate the average price and round to 2 decimal places
avg_price = round(merged_df['price_cleaned'].mean(), 2)

#print the result
print(f"Average listing price: ${avg_price}")
```

## Creating a DataFrame with the Four Solution Values

```python
df_review_dates = pd.DataFrame({
    'first_reviewed': [first_reviewed],
    'last_reviewed': [last_reviewed],
    'nb_private_rooms': [nb_private_rooms],
    'avg_price': [avg_price]
})

# Display the result
print(df_review_dates)
```
